```python
import numpy as np
import pandas as pd
from astropy.io import ascii
import os
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = ascii.read("ngc5288_gaia_cutquality.csv") 
```

```python
pmra = data["pmra"]
pmde = data["pmdec"]
gmag = data["phot_g_mean_mag"]
plx  = data["plx"] 
ra  = data["ra"]
dec = data["dec"]
```

#SIMBAD
#pmRA = -7.422, pmDE =  -5.534 


```python
#limiting the proper motion +-5 from SIMBAD values
kk, = np.where(((pmra > -12.42)& (pmra < -2.42))&((pmde > -10.53)&(pmde < -0.53)))
```

```python
#kk, = np.where((plx>0.5)&(plx<0.7))
```

```python
data2 = data[kk]

pmra2 = data2["pmra"]
pmde2 = data2["pmdec"]
gmag2 = data2["phot_g_mean_mag"]
plx2  = data2["plx"] 
ra2   = data2["ra"]
dec2  = data2["dec"]
```

```python
#%matplotlib notebook

fig, ax = plt.subplots()

ax.scatter(pmra, pmde,color='gray', s=8)
ax.scatter(pmra2, pmde2,color='blue',alpha=0.1, s=8)
ax.set_xlim(-15,2)
ax.set_ylim(-15,2)
ax.set_ylabel("pmDE")
ax.set_xlabel("pmRA")
```

```python
#%matplotlib notebook

fig, ax = plt.subplots()

#ax.scatter(pmra, pmde,color='gray', s=8)
ax.scatter(pmra2, pmde2,color='blue',alpha=0.1, s=8)
ax.scatter(-7.422,-5.534, s=15, color='r')
ax.set_xlim(-8,-5)
ax.set_ylim(-6,-4)
ax.set_ylabel("pmDE")
ax.set_xlabel("pmRA")
```

```python
#%matplotlib notebook

fig, ax = plt.subplots()

ax.scatter(plx, gmag,color='gray', s=8)
ax.scatter(plx2, gmag2,color='blue',alpha=0.5, s=8)
ax.set_ylabel("G")
ax.set_xlabel("Plx")
ax.set_xlim(0,5)
ax.set_ylim(16,6)
```

```python
%matplotlib notebook

fig, ax = plt.subplots()

ax.scatter(ra, dec, color='gray', s=8)
ax.scatter(ra2, dec2,alpha=0.1,color='blue', s=8)
ax.set_ylabel("Dec")
ax.set_xlabel("RA")
```
